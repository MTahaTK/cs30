# Python Standard Library

"The Python Standard Library contains a huge number of useful modules and is part of every standard Python installation. It is important to become familiar with the Python Standard Library since many problems can be solved quickly if you are familiar with the range of things that these libraries can do." H, Swaroop C. “Standard Library.” Data Structures · A Byte of Python, [python.swaroopch.com/stdlib.html](https://python.swaroopch.com/stdlib.html)

There are certain features of Python can only be accessed if you state in your code that you want to import them. By only access the commands and characteristics you need, it allows your code to run faster. Consider the following example:

```
import math

r = float(input("What is your circle's radius in cm? "))

area_circle = math.pow(r, 2) * math.pi

print("The circle has an area of " + str(area_circle) + " cm^2")
```

A note about the above example, math.pow(x, y), functions similarly to the ** operator. Unlike the built-in ** operator, math.pow(x, y) converts both its arguments to type float.

By importing the math module, we are able to access constants like pi, operators like square root, and things like the trigonometric functions.
